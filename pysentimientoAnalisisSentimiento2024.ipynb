```python
from pysentimiento import create_analyzer
import transformers

transformers.logging.set_verbosity(transformers.logging.ERROR)

analyzer = create_analyzer(task="sentiment", lang="es")
```

```python
# Formato de fecha admitido: 04/12/2024 o 4/12/2024

# Programa 1

# Formateo de los datos para aceptar comillas dobles


import re

def procesar_archivo(nombre_archivo):
    # Expresión regular para buscar fechas en formato dd/mm/aaaa
    patron_fecha = r'(([0-2][0-9]|3[0-1]|[1-9])\/\d{2}\/\d{4})'
    # Expresión regular para buscar comillas
    patron_comillas = r'"'

    # Diccionario para almacenar la información
    datos = {}
    fecha_actual = None
    oraciones = []

    with open(nombre_archivo, 'r') as archivo:
        for linea in archivo:
            # Omitir renglones vacíos
            if not linea.strip():
                continue

            # Buscar fecha en la línea
            coincidencias_fecha = re.findall(patron_fecha, linea)
            if coincidencias_fecha:
                # Si se encuentra una fecha, actualizar fecha actual y guardar oraciones anteriores
                if fecha_actual and oraciones:
                    datos[fecha_actual] = oraciones
                fecha_actual = coincidencias_fecha[0][0]  # Obtener solo la fecha completa
                oraciones = []

            # Buscar comillas en la línea y agregar símbolo '\' antes de ellas
            linea = re.sub(patron_comillas, r'\\"', linea)

            # Si la línea no contiene una fecha, agregarla al listado de oraciones
            if not coincidencias_fecha:
                oraciones.append(linea.strip())

        # Guardar las últimas oraciones
        if fecha_actual and oraciones:
            datos[fecha_actual] = oraciones

    return datos

# Ejemplo de uso
nombre_archivo = "encabezados_universal_x.txt"
#nombre_archivo = "encabezados_sub_universal.txt"
#nombre_archivo = "encabezados_sub_jornada.txt"
datos = procesar_archivo(nombre_archivo)
for fecha, oraciones in datos.items():
    print(f"Fecha: {fecha}")
    print("Oraciones:")
    for oracion in oraciones:
        print(oracion)
    print()
```

```python
# Programa 2

# Programa de prueba para determinar los sentimientos de cada encabezado por fecha
# Imprime los datos en consola

def contar_sentimientos_por_fecha(diccionario):
    contador_sentimientos = {"POS": 0, "NEU": 0, "NEG": 0}

    for fecha, oraciones in diccionario.items():
        print(f"Fecha: {fecha}")
        contador_sentimientos = {"POS": 0, "NEU": 0, "NEG": 0}
        for oracion in oraciones:
            resultado = analyzer.predict(oracion)
            sentimiento = resultado.output
            #print(sentimiento)
            # Incrementar el contador correspondiente
            if sentimiento == "POS":
                contador_sentimientos["POS"] += 1
            elif sentimiento == "NEU":
                contador_sentimientos["NEU"] += 1
            elif sentimiento == "NEG":
                contador_sentimientos["NEG"] += 1

        print("Cantidad de oraciones por sentimiento:")
        for sentimiento, cantidad in contador_sentimientos.items():
            print(f"{sentimiento}: {cantidad}")

        print()

# Suponiendo que 'datos' es el diccionario generado en el programa anterior
contar_sentimientos_por_fecha(datos)
```

```python
# Programa DEFINITIVO para calcular el Cohen's Kappa a partir de un CSV
# con columnas denominadas "opinion1" y "opinion2"

import csv
from sklearn.metrics import cohen_kappa_score

def leer_csv(nombre_archivo):
    columnas = {}  # Diccionario para almacenar las columnas

    with open(nombre_archivo, newline='') as archivo_csv:
        lector_csv = csv.reader(archivo_csv)

        # Leer la primera fila para obtener los nombres de las columnas
        nombres_columnas = next(lector_csv)

        # Inicializar las listas vacías para cada columna
        for columna in nombres_columnas:
            columnas[columna] = []

        # Leer el resto del archivo y almacenar los datos en las listas correspondientes
        for fila in lector_csv:
            for i, valor in enumerate(fila):
                nombre_columna = nombres_columnas[i]
                columnas[nombre_columna].append(valor)

    return columnas

# Ejemplo de uso
#nombre_archivo = 'jornada_info_CKappa.csv'
#nombre_archivo = 'universal_info_CKappa.csv'
#nombre_archivo = 'milenio_info_CKappa.csv'
#nombre_archivo = 'jornada_sent_CKappa.csv'
#nombre_archivo = 'universal_sent_CKappa.csv'
nombre_archivo = 'milenio_sent_CKappa.csv'

datos = leer_csv(nombre_archivo)

# Mostrar los datos en arrays
for columna, valores in datos.items():
    print(f"{columna}: {valores}")


# Se crean dos vectores de opiniones

opinion1, opinion2 = [], []

# Se copia el contenido de dos arreglos dentro de un diccionario
# Se establece el nombre de las columnas del archivo .csv
opinion1 = datos['\ufeffopinion1'][:]
opinion2 = datos['opinion2'][:]

# Se calcula el Kappa de Cohen
cohen_kappa_score(opinion1,opinion2)



```
